```python
name=[]
year=[]
mode=[]
model=[]
km_range=[]
user_no=[]
fuel=[]
pay_per_month=[]
total_price=[]
downpayement=[]


for i in range(1,50):
    for result in results:
        try:
            name.append(result.find('h2').get_text())
        except:
            name.append('n/a')
        
        
        try:
            my_list2=result.find('h2').get_text().split(" ")
            x=len(my_list2)
            year.append(my_list2[0])
        except:
            year.append('n/a')    
        
        try:
            my_list=result.find('p', {'class': 'cvakB'}).get_text().split(' ')
            x=len(my_list)
            mode.append(my_list[x-1])
        except:
            mode.append('n/a')
            
        try:
            my_list1=result.find('ul', {'class': 'bVR0c'}).get_text().split(' ')
            z=len(my_list1)
            y=my_list1[0]
            km_range.append("".join(y))
            
        except:
            km_range.append('n/a')
               
                
        try:
            my_list1=result.find('ul', {'class': 'bVR0c'}).get_text().split(' ')
            z=len(my_list1)
            fuel.append(my_list1[z-1])
        except:
            fuel.append('n/a')
            
            
        try:
            my_list1=result.find('ul', {'class': 'bVR0c'}).get_text().split(' ')
            z=len(my_list1)
            user_no.append(my_list1[1])
        except:
            user_no.append('n/a')   
        
                    
                    
                    
        try:
            pay_per_month.append(result.find('div', {'class': '_2HFRN'}).get_text())
        except:
            pay_per_month.append('n/a')
                    
                    
                    
                    
        try:
            total_price.append(result.find('div', {'class': '_7udZZ'}).get_text())
        except:
            total_price.append('n/a')
                    
                    
       
    
        try:
               downpayement.append(result.find('div', {'class': 'JvSRy'}).get_text())
        except:
               downpayement.append('n/a')
                
                
        i=i+1
```

```python
name
```

```python
car_dealer=pd.DataFrame({'Name':name, 'Year':year,'Mode': mode,'kilometers covered':km_range, 'Fuel used': fuel, 'number of people the car has been used by': user_no, 'amount to be paid every month': pay_per_month, 'total amount to be paid': total_price, 'downpayement': downpayement })
```

```python
car_dealer
```

```python
car_dealer['amount to be paid every month']=car_dealer['amount to be paid every month'].apply(lambda x:x.strip('/month'))
```

```python
car_dealer['Fuel used']=car_dealer['Fuel used'].apply(lambda x:x.strip('Owner'))
```

```python
car_dealer['number of people the car has been used by']=car_dealer['number of people the car has been used by'].apply(lambda x:x.strip('km'))
car_dealer['number of people the car has been used by']=car_dealer['number of people the car has been used by'].apply(lambda x:x.strip('st'))
car_dealer['number of people the car has been used by']=car_dealer['number of people the car has been used by'].apply(lambda x:x.strip('nd'))
car_dealer['number of people the car has been used by']=car_dealer['number of people the car has been used by'].apply(lambda x:x.strip('rd'))
```

```python
car_dealer['amount to be paid every month']=car_dealer['amount to be paid every month'].apply(lambda x:x.strip('₹'))
car_dealer['amount to be paid every month']=car_dealer['amount to be paid every month'].apply(lambda x:x.replace(',', ''))
```

```python
car_dealer['total amount to be paid']=car_dealer['total amount to be paid'].apply(lambda x:x.strip('₹'))
car_dealer['total amount to be paid']=car_dealer['total amount to be paid'].apply(lambda x:x.replace(',', ''))
```

```python
car_dealer['kilometers covered']=car_dealer['kilometers covered'].apply(lambda x:x.replace(',', ''))
```
